```julia
using MCTS
using POMDPModels

function MCTS.node_tag(s::GridWorldState)
    if s.done
        return "done"
    else
        return "[$(s.x),$(s.y)]"
    end
end

MCTS.node_tag(a::GridWorldAction) = "[$(a.direction)]"
```

```julia
n_iter = 10000
depth = 10
ec = 10.0

solver = DPWSolver(n_iterations=n_iter, depth=depth, exploration_constant=ec)
mdp = GridWorld()

policy = solve(solver, mdp)

state = GridWorldState(9,1)

a = action(policy, state)
```
